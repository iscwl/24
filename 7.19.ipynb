```python
import random
answer=random.randint(1,100)
#玩家输入数值
n=int(input("Pleasee input num (1-100):"))
#判断输入数字大小
while n!=answer:
    if n>answer:
       n = int(input("Num is more Pleasee continue input num:"))
    elif n<answer:
       n = int(input("Num is less Pleasee continue input num:"))
#输入正确
print("You win the game")
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            print(i,j,k)
```
